```python
bases.columns = ['Dispatching_base_number', 'Company']
```

```python
df['Dispatching_base_number'] = df['Dispatching_base_number'].apply(str)
bases['Dispatching_base_number'] = bases['Dispatching_base_number'].apply(str)
df['Dispatching_base_number'] = df['Dispatching_base_number'].replace(' ', '')
bases['Dispatching_base_number'] = bases['Dispatching_base_number'].replace(' ','')
```

```python
final = df.merge(bases, on = 'Dispatching_base_number', how = 'left')
```

```python
bk = final[final['Dispatching_base_number'] == 'B00037']
```

```python
bk['Pickup_DateTime'] = pd.to_datetime(bk['Pickup_DateTime'])
bk['DropOff_datetime'] = pd.to_datetime(bk['DropOff_datetime'])
bk['Pickup_DateTime'] = bk['Pickup_DateTime'].apply(lambda x : (x-datetime.datetime(1970,1,1)).total_seconds())
bk['DropOff_datetime'] = bk['DropOff_datetime'].apply(lambda x : (x-datetime.datetime(1970,1,1)).total_seconds())
```

```python
bk['total_time'] = bk['DropOff_datetime'] - bk['Pickup_DateTime']
```
